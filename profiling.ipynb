### `_get_info_safe` (0.108s)

Split evenly between two sub-calls:

**`_get_variable` (0.052s)**
- `_arrange_dims` (0.026s): dominated by `RasterArray.write_crs` → xarray coordinate alignment/reindexing (~0.018s)
- Second `write_crs` call (0.016s): `CRS.to_cf` + `pyproj` overhead
- `DataTree.__getitem__` (0.009s): navigating the tree + `calculate_dimensions`

**`XarrayReader.__init__` (0.052s)**
- `RasterArray.bounds` (0.026s) and `RasterArray.transform` (0.018s) both repeatedly call `_internal_bounds` → `DataArray.__getitem__` → `IndexVariable.isel`
- **Redundancy:** `_internal_bounds` is called ~6 times across `bounds`, `transform`, and `resolution` — all hitting the same coordinate indexing logic

### Key Observations

1. **`write_crs` called twice** in `_get_variable` — once in `_arrange_dims` and once after. The second call re-does pyproj `CRS.to_cf` which isn't cheap.
2. **`_internal_bounds` called repeatedly** (~6x) — `bounds`, `transform`, and `resolution` each recompute it independently rather than sharing results.
3. No S3 I/O — all local/cached operations, so this is pure CPU/Python overhead.

### Bottleneck: actual data fetch (0.480s / 92%)

`XarrayReader.preview` → `DataArray.to_masked_array` → same lazy materialization chain:

```
MemoryCachedArray → CopyOnWriteArray → LazilyIndexedArray
  → ZarrArrayWrapper.__getitem__
    → zarr Array.get_orthogonal_selection
      → sync() → lock.acquire  ← S3 pixel data read (0.479s)
```

This is **expected** — the dominant cost is legitimate I/O fetching the preview pixels. The remaining ~0.04s of Python overhead is reasonable.

### Secondary cost: `__init__` overhead (0.026s)

Same repeated pattern: `_get_groups` → `_validate_zarr` → `RasterDataset.crs` → `_get_spatial_dims` on every reader instantiation.

### Key Observations

The ~0.48s is the irreducible cost of fetching real data from S3. The remaining ~0.04s is pure Python overhead (reader init, _get_variable, etc.), which is reasonable. This cell looks healthy — the dominant cost is legitimate I/O, not wasted work.
